```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Câu 1: 
Trong dữ liệu có những đặc điểm nào của xe? Những đặc điểm đó có tương quan với giá xe không?

```python
df = pd.read_csv("hw7_cars.csv")
df.isna().sum()
```

- Đặc điểm của xe (ngoại trừ Price): Quãng đường đã đi, tuổi xe, năm tạo ra, model xe

```python
numeric_var_corr = df.corr(numeric_only=True)['Price'].sort_values()
numeric_var_corr
```

- Quãng đường xe đã đi càng dài, giá càng thấp
- Tuổi càng cao giá xe càng thấp
- Năm tạo ra càng mới/cao, giá càng cao

### Câu 2:
Nếu có, hãy vẽ scatterplot biểu diễn mối quan hệ giữa biến có hệ số tương quan cao nhất với giá xe.

```python
sns.scatterplot(x=df['year'], y=df['Price'], label='Dữ liệu thực tế')
plt.xlabel('Year created')
plt.ylabel('Price')
plt.title('Mối quan hệ giữa năm được tạo và giá xe')
plt.show()
```

### Câu 3
Xây dựng mô hình hồi quy tuyển tính đơn biến dự đoán giá xe. Sau đó vẽ đường hồi quy tính toán được trên scatterplot ở câu 2.

```python
X = df[['year']]
y = df['Price']
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
sns.scatterplot(x=df['year'], y=df['Price'], label='Dữ liệu thực tế')
plt.plot(X, y_pred, color='orange')
plt.xlabel('Year created')
plt.ylabel('Price')
plt.title('Mối quan hệ giữa năm được tạo và giá xe')
plt.show()
```
