```python
import pandas as pd
df = pd.read_csv("Billionaires.csv")
df.head()
```

# Xử lí dữ liệu NaN
- isnull().sum()
- dropna(axis=1): `xóa tất cả các cột có chứa dữ liệu null`
- drop([`tencot`], axis=1, inplace=True): `bỏ các cột không cần thiết`

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
#hàm ghi đè lên dữ liệu cũ
df.drop(["organization"], axis=1, inplace=True)
```

# Điểm khuyết
- fillna()
- bfill: `thay giá trị bằng giá trị ngay sau`
- ffill: `thay giá trị khuyết bằng giá trị ngay trước nó`

```python
nan_rows_any = df[df.isna().any(axis=1)]

print("\nRows with at least one NaN value:")
print(nan_rows_any)
```

```python
df_fill_0 = df.fillna(0)
df_fill_0.isnull().sum()
df_fill_0.tail()
```

```python
df_bfill = df.bfill()
df_bfill.isnull().sum()
df_bfill.tail()
```

```python
df_ffill = df.ffill()
df_ffill.isnull().sum()
```

# Ép kiểu
- dtypes
- Series.astype('<kiểu dữ liệu>')
- pd.to_numeric(<cột>): để pandas chọn kiểu phù hợp

```python
df.dtypes
```

```python
df_ffill.age = df_ffill.age.astype("int64")
df_ffill.age.dtype
```

```python
df_ffill.birthYear = pd.to_numeric(df_ffill.birthYear)
df_ffill.birthYear.dtype
```
