```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("spotify.csv")
df.isna().sum()
```

```python
df.tail()
```

# Plot
- Biểu đồ đường cong

```python
#bieu do duong cho monthly listener
plt.figure(figsize=(14, 6))
plt.plot(df["Monthly Listeners (Millions)"], marker="o", linestyle="-", linewidth=2, color='red')
plt.ylabel("Monthly Listeners (Millions)", fontsize="14")
plt.title("Biểu đồ đường số người nghe theo tháng")

plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

# bar:
- Biểu đồ cột

```python
#gom nhom data
artist_streams = df.groupby("Artist")["Total Streams (Millions)"].sum().nlargest(10)
artist_streams
```

```python
plt.figure(figsize=(10, 5))
artist_streams.plot(kind="barh", color="#70B2B2", edgecolor="#016B61")

plt.title("Top 10 artists with the highest total streams", fontsize=14)
plt.xlabel("Total streams (mil)", fontsize=12)
plt.ylabel("Artist", fontsize=12)
#dao truc de cao nhat o tren
plt.gca().invert_yaxis()

plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.tight_layout()

plt.show()
```

# Histogram
- Biểu đồ tần suất

```python
df_billi = pd.read_csv("../Lesson4/Billionaires.csv")
df_billi.isnull().sum()
```

```python
#xu li null age
df_billi["age"] = df_billi["age"].fillna(df_billi["age"].mean().round())
df_billi.isnull().sum()
```

```python
#doi kieu int cho age
df_billi["age"] = df_billi["age"].astype("int")
df_billi["age"]
```

```python
plt.hist(x=df_billi["age"], color="#70B2B2", edgecolor="#016B61")
plt.title("Biểu đồ cột phân bố tuổi")
plt.ylabel("Số người")
plt.xlabel("Số tuổi")
plt.show()
```
